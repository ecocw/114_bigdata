## 2.資料轉換—離散轉連續／連續轉離散
### (1) 離散轉連續：獨熱編碼 (One-Hot Encoding)
pandas.get_dummies用法¶

```python
import pandas as pd      #匯入Pandas套件，命名縮寫為pd
file='bankwage.csv'    #指定相對路徑
df = pd.read_csv(file)   #讀取csv檔
df.head()                #顯示前5列
```

### (2) 連續資料離散化處理

```python
#刪除有遺漏值的列
df0=df.dropna(axis = 0)  #asix=0 代表row
df0.head() #顯示前5列
```

```python
# df0.describe()
```

#### A.分位數法-使用四分位 pd.qcut()函數¶

```python
#pd.qcut()：依分位數切割數值變數。

wage_quartile=pd.qcut(df0['wage'],    #將wage進行分位數切割
                      q=4,            #切成四等份（四分位數）
                      labels=['Q1', 'Q2', 'Q3', 'Q4']) #給每個區間貼標籤

print(wage_quartile.value_counts())    #列印各區間數目

df0 = df0.copy()  # 避免副本警告
df0['wage_quartile'] = wage_quartile

display(df0)
```

#### B.距離區間法-pd.cut函數¶

```python
bins_wage = [0, 30000, 60000, 90000, 120000, 150000]   #指定多區間
wage_interval =pd.cut(df0['wage'], bins_wage)          # 使用cut方法進行分箱
print(wage_interval.value_counts())         # 統計每個區間人數

df0 = df0.copy()  # 避免副本警告
df0['wage_quartile'] = pd.cut(df0['wage'],bins_wage)
display(df0)
```

## 3. 資料轉換—標準化與正規化
### (1) Z score 標準化

```python
import pandas as pd
from sklearn import preprocessing

train=df0[['wage']]
wage_zscore=preprocessing.scale(train, axis=0,
                           with_mean=True,             #利用平均數
                           with_std=True, copy=True)   #利用標準差

df0.loc[df0.index,'wage_zscore']= wage_zscore
display(df0.head())
```

### (2) min-max正規化

```python
from sklearn.preprocessing import MinMaxScaler

# 創建MinMaxScaler
scaler = MinMaxScaler()

# 對欄位進行轉換
wage_normalized = scaler.fit_transform(df0[['wage']])

#加入df0
df0.loc[df0.index, 'wage_normalized']= wage_normalized
display(df0.head())
```

## 5.資料整合
### (1) Append (縱向追加)

```python
import pandas as pd      #匯入Pandas套件，命名縮寫為pd

employee=['Bob','Jake','Lisa','Sue']
group=['Accounting','Engineering','Engineering','HR']
#創建字典
dict={'employee':employee,'group':group}

#轉為DataFrame
df1=pd.DataFrame(dict)
display(df1)                #顯示

import pandas as pd      #匯入Pandas套件，命名縮寫為pd
employee=['Eliza','Jerry']
group=['Finance','HR']
df2=pd.DataFrame({'employee':employee,'group':group})
display(df2)                #顯示

#縱向追加
df3=pd.concat([df1, df2], ignore_index=True)#索引重新編碼
display(df3)               #顯示
```

### (2) Merge(基於指定行的合併拼接)¶

```python
import pandas as pd      #匯入Pandas套件，命名縮寫為pd
employee1=['Bob','Jake','Lisa','Sue']
group1=['Accounting','Engineering','Engineering','HR']
df1=pd.DataFrame({'employee':employee1,'group':group1})
display(df1)                #顯示

employee2=['Lisa','Bob','Jake','Sue']
hire_date=[2004,2008,2012,2014]
df2=pd.DataFrame({'employee':employee2,'hire_date':hire_date})
display(df2)                #顯示

df3=pd.merge(df1,df2)
display(df3)
```

### (3) Concat (資料串接)

```python
#利用字典創造DataFrame
dict1={'a':[1,2],'b':[2,4]}
df1 = pd.DataFrame(dict1,index=[1, 2]) #若未指定index, 由0開始
display(df1)

#利用字典創造DataFrame
dict2={'c':[3,4],'d':[6,8]}
df2 = pd.DataFrame(dict2,index=[2, 3]) #若未指定index, 由0開始
display(df2)

#橫向合併-inner
df_concat1=pd.concat([df1,df2],axis=1, join='inner') #列交集
display(df_concat1)

#橫向合併-outer
df_concat2=pd.concat([df1,df2],axis=1, join='outer') #列聯集
display(df_concat2)
```

### (4) Join (基於列索引進行拼接)

```python
courses=['Spark','PySpark','Python','pandas']
fee=[20000,25000,22000,30000]
duration=['30days','40days','35days','50days']
dict={'Cources':courses,'Fee':fee,'Duration':duration}
df1=pd.DataFrame(dict,index=['r1','r2','r3','r4'])
display(df1)
```

```python
courses=['Spark','Java','Python','Go']
discount=[2000,2300,1200,2000]
dict={'Cources':courses,'Discount':discount}
df2=pd.DataFrame(dict,index=['r1','r6','r3','r5'])
display(df2)
```

```python
# pandas left join
df3=df1.join(df2,lsuffix="_left",rsuffix="_right", how='left')
display(df3)
```

```python
#pandas right join
df3=df1.join(df2,lsuffix="_left",rsuffix="_right", how='right')
display(df3)
```

```python
#pandas inner join
df3=df1.join(df2,lsuffix="_left",rsuffix="_right", how='inner')
display(df3)
```

```python
#pandas outer join
df3=df1.join(df2,lsuffix="_left",rsuffix="_right", how='outer')
display(df3)
```
