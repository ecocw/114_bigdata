## 2.網頁爬蟲

### (1) 使用request.get()套件
    匯入requests套件, 可以在ANACONDA PROMPT下輸入 pip3 install requests     
    requests.get()模擬HTTP GET 方法送出請求至遠方的伺服器,伺服器回傳網頁原始碼
### (2) BeautifulSoup 套件 整理成結構化資料
    匯入BeautifulSoup套件, 在ANACONDA PROMPT下輸入 pip3 install BeautifulSoup4
    BeautifulSoup 的運作方式就是讀取 HTML 原始碼，自動進行解析並產生一個 BeautifulSoup 物件，
    接下來所有資料的搜尋、萃取等操作都會透過這個物件來進行.

```python
import requests
from bs4 import BeautifulSoup

# PTT 板網址
url = 'https://www.ptt.cc/bbs/Finance/index.html'

# 模擬HTTP GET 方法送出請求至伺服器回傳網頁原始碼
res = requests.get(url)

# 設定編碼
res.encoding = 'utf-8'

# 使用 BeautifulSoup 解析 HTML文字檔 res.text
soup = BeautifulSoup(res.text, 'html.parser')

# 找出所有文章標題 soup.select('div.title a') → 找到文章標題與超連結
titles = soup.select('div.title a')  # <div class="title"><a href=...>文章標題</a></div>

# 輸出結果  title.text.strip() → 取得標題文字 title['href'] → 取得文章網址，再補上 https://www.ptt.cc。
for i, title in enumerate(titles, 1):
    print(f"{i}. {title.text.strip()} - https://www.ptt.cc{title['href']}")
```

```python
import requests
from bs4 import BeautifulSoup

# PTT板網址
url = 'https://www.ptt.cc/bbs/gossiping/index.html'

# 模擬HTTP GET 方法送出請求至伺服器回傳網頁原始碼存成res
# 加上 over18 cookie，跳過年齡驗證
res = requests.get(url, cookies={'over18': '1'})

# 設定編碼
res.encoding = 'utf-8'

# 使用 BeautifulSoup 解析 HTML
soup = BeautifulSoup(res.text, 'html.parser')

# 找出所有文章標題 soup.select('div.title a') → 找到文章標題與超連結
titles = soup.select('div.title a')  # <div class="title"><a href=...>文章標題</a></div>

# 輸出結果  title.text.strip() → 取得標題文字 title['href'] → 取得文章網址，再補上 https://www.ptt.cc。
for i, title in enumerate(titles, 1):
    print(f"{i}. {title.text.strip()} - https://www.ptt.cc{title['href']}")
```

### (1) 範例: 到yahoo finance 抓取調整收盤價

```python
import pandas as pd        # 導入 pandas 套件
import yfinance as yf      # 導入 yfinance 套件
from datetime import datetime

# 設定股票代碼
tickers = ['META', 'AAPL', 'AMZN', 'NFLX', 'GOOGL']

# 設定日期區間
start_date = datetime(2020, 1, 1)
end_date = datetime(2024, 12, 31)

# 下載股票資料 (自動調整收盤價)
data = yf.download(tickers, start=start_date, end=end_date, auto_adjust=True)

# 顯示前五筆
display(data.head())

# 只取出收盤價 (Close)
close_prices = data['Close']

# 顯示前五筆
display(close_prices.head())
```

### (2) 範例：上網抓取多檔股票歷史日收盤價

```python
#上網路爬取股價資料
import pandas as pd
import yfinance as yf

# 下載資料起始日與股票代號
stockid = ['2303.TW', '2330.TW', '3008.TW', '2498.TW']
start = '2020-01-01'
end = '2025-08-31'

# 使用 ExcelWriter (指定 openpyxl 引擎)
with pd.ExcelWriter('stockprice.xlsx', engine='openpyxl') as writer:
    # 依公司代號順序抓取資料
    for sid in stockid:
        df = yf.download(sid, start=start, end=end, auto_adjust=True)
        df.to_excel(writer, sheet_name=sid)  # 各公司一個工作表
```
